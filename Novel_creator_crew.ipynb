```python
plot='lucas-a high school playboy falls for maya- a college going nerd.'
```

```python
world = Task(
    description=f""" Craft an intricate fictional world  and add details about geography, complex history if necessary for the given plot: {plot}.""",
    agent=Worldbuilder,
    expected_output="a detailed description of a fictional world"
)

characters = Task(
    description=f"""Craft a compelling cast of characters for the plot {plot}, each with rich backstories, distinct personalities, intricate relationships, and authentic dialogue reflecting their unique voices and emotions. Ensure the presence of supporting characters who play significant roles alongside the main characters in the novel.""",
    agent=CharacterCrafter,
    expected_output="character profiles for the novel, detailing backstory, motivations, and goals, along with fleshed-out personalities, defined relationships, and sample dialogue snippets showcasing their emotional range in 400-500 words"
)

Plot = Task(
    description=f"""Craft an engaging narrative structure by creating a central conflict, escalating stakes, defining clear plot points, and developing a satisfying resolution. Incorporate genre conventions, character arcs, well-timed plot twists, and balanced pacing to maintain intrigue and tension for the following plot : {plot} and characters:{characters}""",
    agent=Plotarchitect,
    expected_output="narrative structure for the novel, including the central conflict, escalating challenges, key plot points, satisfying resolution, genre-specific elements, character arcs, plot twists, foreshadowing, and pacing plan for a satisfying rhythm in 500 words"
)

initial_stry = Task(
    description=f"""Write the prologue ,introducing the novel to its readers, use the world built:{world} by the Worldbuilder agent,the characters written :{characters}by the CharacterCrafter, to create this. Make sure that the epilogue gives only the hint and makes the reader want to read the whole story. Don't disclose the whole novel in the prologue,just give a intro of the characters.""",
    agent=writer,
    expected_output="A prologue introducing the story to the readers in about 1000 words."
)

```

```python
crew = Crew(
    agents=[Worldbuilder,CharacterCrafter,Plotarchitect,writer],
    tasks=[world,characters,Plot,initial_stry],
    verbose=4
)
```

```python
result = crew.kickoff()

print("------------------------------------------------NOVEL----------------------------------------------------")
result
```
